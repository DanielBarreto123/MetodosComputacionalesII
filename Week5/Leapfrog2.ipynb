```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega = 2*np.pi
```

```python
def f(x,t,omega):
    return -omega**2*x
```

```python
def LeapFrog(f,x0,t,omega):
    
    h = t[1] - t[0]
    
    print(h)
    
    x = np.zeros_like(t)
    v = np.zeros_like(t)
    
    x[0] = x0[0]
    v[0] = x0[1]
    
    for i in range(1,len(t)):
        vhalf = v[i-1] + 0.5*h*f(x[i-1],t[i-1],omega)
        
        x[i] = x[i-1] + h*vhalf
        
        v[i] = vhalf + 0.5*h*f(x[i],t[i],omega)
        
    return x,v
```

```python
t = np.linspace(0,5,100)
x0 = np.array([1,0])
sol = LeapFrog(f,x0,t,omega)
```

```python
plt.plot(t,sol[0])
```

```python
plt.plot(sol[0],sol[1])
```
