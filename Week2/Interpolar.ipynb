```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1,2,3])
Y = np.array([4,6,10])
```

```python
plt.scatter(X,Y)
```

```python
def Lagrange(x,X,i):
    
    L = 1.
    
    for j in range(X.shape[0]):
        if j != i:
            L *= (x - X[j])/(X[i]-X[j])
        
    return L
```

```python
def Interpolate(x,X,Y):
    
    Poly = 0.
    
    for i in range(X.shape[0]):
        Poly += Y[i]*Lagrange(x,X,i)
        
    return Poly
```

```python
Lagrange(0.3,X,0)
```

```python
x = np.linspace(0,4,10)
Polinomio = Interpolate(x,X,Y)
```

```python
plt.scatter(X,Y,color='r')
plt.plot(x,Polinomio)
```
