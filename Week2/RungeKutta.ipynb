```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
```

```python
def f(x,y):
    return x + 2*y
```

```python
def Solucion(x):
    return -0.5*x - (1-np.exp(2*x))/4.
```

```python
N = 100
t = np.linspace(0,1,N)
t
```

```python
h = t[1] - t[0]
```

```python
yext = Solucion(t)
```

```python
plt.plot(t,yext)
```

```python
def GetEuler(x,y,h):
    
    for i in range(1,len(x)):
        
        k1 = f(x[i-1],y[i-1])
        y[i] = y[i-1] + h*k1
```

```python
y1 = np.zeros_like(t)
y1[0] = 0
```

```python
GetEuler(t,y1,h)
```

```python
plt.plot(t,yext)
plt.plot(t,y1)
```
