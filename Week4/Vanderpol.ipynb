```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def System(r,t,mu=1.5):
    
    x,vx = r
    
    dxdt = vx
    dvxdt = mu*(1-x**2)*vx - x
    
    return [dxdt,dvxdt]
```

```python
t = np.linspace(0,20,10000)
r0 = [0.5,0.]
```

```python
sol = odeint(System, r0, t, args=(1.3,))
```

```python
plt.plot(sol[:,0],sol[:,1])
```
