```python
!rm Data/*
file = 'Data/data.dat'
url = 'https://raw.githubusercontent.com/asegura4488/Database/main/MetodosComputacionalesReforma/Sismo.txt'
if not path.exists(file):
    Path_ = wget.download(url,file)
    print('File loaded')
else:
    Path_ = file
```

```python
data = np.loadtxt(Path_)
```

```python
a = data[:,1]
```

```python
plt.plot(a)
```

```python
fft = np.fft.fft(a)
```

```python
Npoints = len(a)
Frecuencias = np.fft.fftfreq(Npoints,1)
#Frecuencias = np.fft.fftshift(Frecuencias)
ffta = np.abs(fft)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(Frecuencias,np.abs(fft/Npoints),marker='.')
```

```python
mask = Frecuencias < 0
fft[mask] = 0
```
